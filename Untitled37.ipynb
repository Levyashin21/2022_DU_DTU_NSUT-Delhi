```python
# ============================================
# 0) Install deps (Colab) + Imports
# ============================================
!pip -q install biopython pandas requests scikit-learn umap-learn

import re, time, hashlib
from io import StringIO
from typing import List, Dict

import requests
import numpy as np
import pandas as pd
from Bio import SeqIO

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Optional viz
import umap

sns.set_context("talk")

# ============================================
# 1) UniProt helpers
# ============================================
UNIPROT_FASTA = "https://rest.uniprot.org/uniprotkb/stream"

def fetch_uniprot_fasta(query: str,
                        reviewed: bool = True,
                        tax_filter: str = "(taxonomy_id:2 OR taxonomy_id:2157)",
                        size_limit: int = 10000,
                        retries: int = 3,
                        pause: float = 0.5) -> str:
    """
    Stream FASTA from UniProt REST API for a query.
    - reviewed=True pulls Swiss-Prot; False includes TrEMBL.
    - tax_filter keeps Bacteria/Archaea by default.
    """
    reviewed_q = " AND reviewed:true" if reviewed else ""
    q = f"({query}) AND {tax_filter}{reviewed_q}"
    params = {"query": q, "format": "fasta", "size": size_limit}

    for attempt in range(1, retries+1):
        try:
            r = requests.get(UNIPROT_FASTA, params=params, timeout=60)
            r.raise_for_status()
            text = r.text
            if text.strip():
                return text
            # empty response: backoff and try unreviewed outside
            return ""
        except Exception as e:
            if attempt == retries:
                print(f"[error] UniProt fetch failed after {retries} tries for query: {query}\n{e}")
                return ""
            time.sleep(pause * attempt)
    return ""

def parse_fasta_to_records(fasta_text: str, family_label: str, source_tag: str) -> List[Dict]:
    """Parse UniProt FASTA into a list of dicts with fields: accession, organism, length, sequence, family, source."""
    records = []
    if not fasta_text:
        return records
    for rec in SeqIO.parse(StringIO(fasta_text), "fasta"):
        header = rec.description
        seq = str(rec.seq)
        acc = None
        m = re.search(r"\|([A-Z0-9]+)\|", header)  # >sp|Q99ZW2|...
        if m: acc = m.group(1)
        if acc is None: acc = rec.id

        organism = ""
        mOS = re.search(r" OS=([^=]+?) OX=", header)  # OS=... OX=...
        if mOS:
            organism = mOS.group(1).strip()

        records.append({
            "accession": acc,
            "organism": organism,
            "length": len(seq),
            "sequence": seq,
            "family": family_label,
            "source": source_tag
        })
    return records

# Queries per family (include synonyms)
FAMILY_QUERIES = {
    "Cas9": [
        'protein_name:"CRISPR-associated endonuclease Cas9"',
        'protein_name:"Cas9"'
    ],
    "Cas12": [
        'protein_name:"CRISPR-associated endonuclease Cas12a" OR protein_name:Cpf1',
        'protein_name:"CRISPR-associated endonuclease Cas12b" OR protein_name:C2c1',
        'protein_name:"CRISPR-associated endonuclease Cas12d" OR protein_name:CasY',
        'protein_name:"CRISPR-associated endonuclease Cas12e" OR protein_name:CasX',
        'protein_name:"CRISPR-associated endonuclease Cas12f" OR protein_name:Cas14',
        'protein_name:"Cas12"'
    ],
    "Cas13": [
        'protein_name:"CRISPR-associated ribonuclease Cas13a" OR protein_name:C2c2',
        'protein_name:"CRISPR-associated ribonuclease Cas13b"',
        'protein_name:"CRISPR-associated ribonuclease Cas13c"',
        'protein_name:"CRISPR-associated ribonuclease Cas13d"',
        'protein_name:"Cas13"'
    ],
    "Cas3": [
        'protein_name:"CRISPR-associated endonuclease/helicase Cas3" OR protein_name:Cas3'
    ],
    "Cas10": [
        'protein_name:"CRISPR-associated protein Cas10" OR protein_name:Cas10'
    ]
}

def build_family_records(min_reviewed_target=100) -> pd.DataFrame:
    """
    For each family:
      - pull reviewed entries (Swiss-Prot) across queries
      - if count < min_reviewed_target, augment with unreviewed
    Return DataFrame of all records.
    """
    all_records = []
    for fam, qlist in FAMILY_QUERIES.items():
        seen = set()
        fam_recs = []

        # 1) reviewed
        for q in qlist:
            fasta = fetch_uniprot_fasta(q, reviewed=True)
            recs = parse_fasta_to_records(fasta, fam, source_tag="UniProt_reviewed")
            for r in recs:
                if r["accession"] not in seen:
                    fam_recs.append(r); seen.add(r["accession"])
            time.sleep(0.2)

        # 2) unreviewed (augment)
        if len(fam_recs) < min_reviewed_target:
            for q in qlist:
                fasta = fetch_uniprot_fasta(q, reviewed=False)
                recs = parse_fasta_to_records(fasta, fam, source_tag="UniProt_unreviewed")
                for r in recs:
                    if r["accession"] not in seen:
                        fam_recs.append(r); seen.add(r["accession"])
                time.sleep(0.2)

        print(f"[{fam}] collected: {len(fam_recs)}")
        all_records.extend(fam_recs)

    df = pd.DataFrame(all_records)
    return df

# ============================================
# 2) Build dataset live from UniProt
# ============================================
df = build_family_records()
print("Raw shape:", df.shape)
display(df.head())

# ============================================
# 3) Clean: dedupe + length sanity + family check
# ============================================
if df.empty:
    raise SystemExit("No sequences fetched. Try re-running (temporary network?), or loosen queries.")

# exact-sequence dedup
df["seq_hash"] = df["sequence"].apply(lambda s: hashlib.md5(s.encode()).hexdigest())
before = len(df)
df = df.drop_duplicates(subset="seq_hash").copy()
print(f"Dedup exact sequences: {before} -> {len(df)}")

# length sanity (keeps most Cas effectors; you can adjust)
df = df[(df["length"] >= 200) & (df["length"] <= 2000)].copy()
print("After length filter:", df.shape)

# drop families with too few samples
counts = df["family"].value_counts()
print("Counts by family:\n", counts)
keep = counts[counts >= 30].index.tolist()  # ensure reasonable class sizes
df = df[df["family"].isin(keep)].reset_index(drop=True)
print("After min-size filter:", df.shape, "| families:", sorted(df["family"].unique()))

# save CSV
df.to_csv("cas_uniprot_dataset.csv", index=False)

# ============================================
# 4) Features: amino-acid composition (fast baseline)
# ============================================
AA = "ACDEFGHIKLMNPQRSTVWY"
AA_SET = set(AA)

def aa_comp(seq: str) -> np.ndarray:
    seq = re.sub(r"[^A-Z]", "", seq.upper())
    L = max(len(seq), 1)
    return np.array([seq.count(a)/L for a in AA], dtype=np.float32)

X = np.vstack([aa_comp(s) for s in df["sequence"]])
y = df["family"].values

# ============================================
# 5) Train/test split + baseline model
# ============================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

clf = make_pipeline(
    StandardScaler(),
    LogisticRegression(max_iter=1000, class_weight="balanced")
)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print("\n=== Classification report (AA composition baseline) ===")
print(classification_report(y_test, y_pred))
print("Confusion matrix (labels in alphabetical family order):")
labels_ord = sorted(df["family"].unique())
print(confusion_matrix(y_test, y_pred, labels=labels_ord))

# ============================================
# 6) Quick pattern visuals (optional but nice)
# ============================================
plt.figure(figsize=(9,5))
for fam in labels_ord:
    sns.kdeplot(df.loc[df["family"]==fam, "length"], fill=True, alpha=0.2, label=fam)
plt.xlabel("Protein length (aa)")
plt.title("Length distributions by Cas family")
plt.legend()
plt.tight_layout(); plt.show()

# UMAP on AA composition (just to visualize; embeddings will look better)
X_umap = umap.UMAP(n_neighbors=20, min_dist=0.1, random_state=42).fit_transform(X)
plt.figure(figsize=(7,6))
for fam in labels_ord:
    mask = (y == fam)
    plt.scatter(X_umap[mask,0], X_umap[mask,1], alpha=0.7, label=fam)
plt.title("UMAP (AA composition features)")
plt.legend()
plt.tight_layout(); plt.show()

print("\nSaved dataset to cas_uniprot_dataset.csv")
```

```python
# ============================================
# 0) Install deps (Colab) + Imports
# ============================================
!pip -q install biopython pandas requests scikit-learn umap-learn xgboost shap

import re, time, hashlib, itertools
from io import StringIO
from typing import List, Dict

import requests
import numpy as np
import pandas as pd
from Bio import SeqIO

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

import umap
import xgboost as xgb
import shap

sns.set_context("talk")

# ============================================
# 1) UniProt helpers
# ============================================
UNIPROT_FASTA = "https://rest.uniprot.org/uniprotkb/stream"

def fetch_uniprot_fasta(query: str,
                        reviewed: bool = True,
                        tax_filter: str = "(taxonomy_id:2 OR taxonomy_id:2157)",
                        size_limit: int = 10000,
                        retries: int = 3,
                        pause: float = 0.5) -> str:
    reviewed_q = " AND reviewed:true" if reviewed else ""
    q = f"({query}) AND {tax_filter}{reviewed_q}"
    params = {"query": q, "format": "fasta", "size": size_limit}
    for attempt in range(1, retries+1):
        try:
            r = requests.get(UNIPROT_FASTA, params=params, timeout=60)
            r.raise_for_status()
            text = r.text
            if text.strip():
                return text
            return ""
        except Exception as e:
            if attempt == retries:
                print(f"[error] UniProt fetch failed after {retries} tries for query: {query}\n{e}")
                return ""
            time.sleep(pause * attempt)
    return ""

def parse_fasta_to_records(fasta_text: str, family_label: str, source_tag: str) -> List[Dict]:
    records = []
    if not fasta_text:
        return records
    for rec in SeqIO.parse(StringIO(fasta_text), "fasta"):
        header = rec.description
        seq = str(rec.seq)
        m = re.search(r"\|([A-Z0-9]+)\|", header)
        acc = m.group(1) if m else rec.id
        mOS = re.search(r" OS=([^=]+?) OX=", header)
        organism = mOS.group(1).strip() if mOS else ""
        records.append({
            "accession": acc,
            "organism": organism,
            "length": len(seq),
            "sequence": seq,
            "family": family_label,
            "source": source_tag
        })
    return records

# Queries per family
FAMILY_QUERIES = {
    "Cas9": [
        'protein_name:"CRISPR-associated endonuclease Cas9"',
        'protein_name:"Cas9"'
    ],
    "Cas12": [
        'protein_name:"CRISPR-associated endonuclease Cas12a" OR protein_name:Cpf1',
        'protein_name:"CRISPR-associated endonuclease Cas12b" OR protein_name:C2c1',
        'protein_name:"CRISPR-associated endonuclease Cas12d" OR protein_name:CasY',
        'protein_name:"CRISPR-associated endonuclease Cas12e" OR protein_name:CasX',
        'protein_name:"CRISPR-associated endonuclease Cas12f" OR protein_name:Cas14',
        'protein_name:"Cas12"'
    ],
    "Cas13": [
        'protein_name:"CRISPR-associated ribonuclease Cas13a" OR protein_name:C2c2',
        'protein_name:"CRISPR-associated ribonuclease Cas13b"',
        'protein_name:"CRISPR-associated ribonuclease Cas13c"',
        'protein_name:"CRISPR-associated ribonuclease Cas13d"',
        'protein_name:"Cas13"'
    ],
    "Cas3": [
        'protein_name:"CRISPR-associated endonuclease/helicase Cas3" OR protein_name:Cas3'
    ],
    "Cas10": [
        'protein_name:"CRISPR-associated protein Cas10" OR protein_name:Cas10'
    ]
}

def build_family_records(min_reviewed_target=100) -> pd.DataFrame:
    all_records = []
    for fam, qlist in FAMILY_QUERIES.items():
        seen, fam_recs = set(), []
        # reviewed first
        for q in qlist:
            fasta = fetch_uniprot_fasta(q, reviewed=True)
            recs = parse_fasta_to_records(fasta, fam, "UniProt_reviewed")
            for r in recs:
                if r["accession"] not in seen:
                    fam_recs.append(r); seen.add(r["accession"])
            time.sleep(0.2)
        # unreviewed if too few
        if len(fam_recs) < min_reviewed_target:
            for q in qlist:
                fasta = fetch_uniprot_fasta(q, reviewed=False)
                recs = parse_fasta_to_records(fasta, fam, "UniProt_unreviewed")
                for r in recs:
                    if r["accession"] not in seen:
                        fam_recs.append(r); seen.add(r["accession"])
                time.sleep(0.2)
        print(f"[{fam}] collected: {len(fam_recs)}")
        all_records.extend(fam_recs)
    return pd.DataFrame(all_records)

# ============================================
# 2) Build dataset
# ============================================
df = build_family_records()
print("Raw shape:", df.shape)
display(df.head())

# Clean
df["seq_hash"] = df["sequence"].apply(lambda s: hashlib.md5(s.encode()).hexdigest())
df = df.drop_duplicates(subset="seq_hash").copy()
df = df[(df["length"] >= 200) & (df["length"] <= 2000)].copy()
counts = df["family"].value_counts()
keep = counts[counts >= 30].index.tolist()
df = df[df["family"].isin(keep)].reset_index(drop=True)
print("After cleaning:", df.shape, "| families:", sorted(df["family"].unique()))
df.to_csv("cas_uniprot_dataset.csv", index=False)

# ============================================
# 3) Features: AA + dipeptides
# ============================================
AA = "ACDEFGHIKLMNPQRSTVWY"
def aa_comp(seq: str) -> np.ndarray:
    seq = re.sub(r"[^A-Z]", "", seq.upper())
    L = max(len(seq), 1)
    return np.array([seq.count(a)/L for a in AA], dtype=np.float32)

DIPEPTIDES = [a+b for a,b in itertools.product(AA, AA)]
def dipeptide_comp(seq: str) -> np.ndarray:
    seq = seq.upper()
    L = max(len(seq)-1, 1)
    counts = {dp:0 for dp in DIPEPTIDES}
    for i in range(len(seq)-1):
        dp = seq[i:i+2]
        if dp in counts:
            counts[dp] += 1
    return np.array([counts[dp]/L for dp in DIPEPTIDES], dtype=np.float32)

X = np.hstack([
    np.vstack([aa_comp(s) for s in df["sequence"]]),
    np.vstack([dipeptide_comp(s) for s in df["sequence"]])
])
y = df["family"].values

print("Feature matrix shape:", X.shape)

# ============================================
# 4) Encode labels
# ============================================
le = LabelEncoder()
y_enc = le.fit_transform(y)
print("Label mapping:", dict(zip(le.classes_, le.transform(le.classes_))))

X_train, X_test, y_train, y_test = train_test_split(
    X, y_enc, stratify=y_enc, test_size=0.2, random_state=42
)

# ============================================
# 5) Train XGBoost
# ============================================
clf_xgb = xgb.XGBClassifier(
    objective="multi:softmax",
    num_class=len(np.unique(y_enc)),
    eval_metric="mlogloss",
    n_estimators=300,
    max_depth=8,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    use_label_encoder=False
)
clf_xgb.fit(X_train, y_train)
y_pred = clf_xgb.predict(X_test)

# ============================================
# 6) Evaluation
# ============================================
print("\n=== Classification report (XGBoost, AA+dipeptides) ===")
print(classification_report(le.inverse_transform(y_test), le.inverse_transform(y_pred)))

print("Confusion matrix:")
print(confusion_matrix(le.inverse_transform(y_test), le.inverse_transform(y_pred), labels=le.classes_))

# ============================================
# 7) Feature importance
# ============================================
feature_names = list(AA) + DIPEPTIDES
imp_df = pd.DataFrame({
    "feature": feature_names,
    "importance": clf_xgb.feature_importances_
}).sort_values("importance", ascending=False)

print("\nTop 20 important features:")
print(imp_df.head(20))

plt.figure(figsize=(10,6))
sns.barplot(x="importance", y="feature", data=imp_df.head(20), palette="viridis")
plt.title("Top 20 important features (XGBoost)")
plt.tight_layout(); plt.show()

# ============================================
# 8) SHAP analysis (per-class interpretability)
# ============================================
explainer = shap.TreeExplainer(clf_xgb)
shap_values = explainer.shap_values(X_test)

print("SHAP output type:", type(shap_values))
if not isinstance(shap_values, list):
    print("SHAP array shape:", shap_values.shape)  # (n_samples, n_features, n_classes)

if isinstance(shap_values, list):
    # older SHAP API: list of arrays
    for i, cls in enumerate(le.classes_):
        print(f"\n=== SHAP importance for {cls} ===")
        shap.summary_plot(
            shap_values[i],
            X_test,
            feature_names=feature_names,
            max_display=15,
            show=False
        )
        plt.title(f"Top features driving {cls} classification")
        plt.show()
else:
    # newer SHAP API: 3D array
    for i, cls in enumerate(le.classes_):
        print(f"\n=== SHAP importance for {cls} ===")
        shap.summary_plot(
            shap_values[:,:,i],
            X_test,
            feature_names=feature_names,
            max_display=15,
            show=False
        )
        plt.title(f"Top features driving {cls} classification")
        plt.show()
```
